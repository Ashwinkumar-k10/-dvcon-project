```python
import os, shutil, zipfile
from google.colab import drive
drive.mount('/content/drive')

DRIVE    = '/content/drive/MyDrive'
VILLAGES = '/content/svamitva/data/train_villages'
SHP_CG   = '/content/svamitva/data/train/shp-file_cg'
SHP_PB   = '/content/svamitva/data/train/shp-file'

# Create folders
for v in ['01','02','03','04','05','06','07','08','09','10']:
    os.makedirs(f'{VILLAGES}/village_{v}', exist_ok=True)
os.makedirs(SHP_CG, exist_ok=True)
os.makedirs(SHP_PB, exist_ok=True)

# All 10 villages in same Drive folder
sources = {
    'village_01': f'{DRIVE}/dataset/village_01_orthophoto.tif',
    'village_02': f'{DRIVE}/dataset/village_02_orthophoto.tif',
    'village_03': f'{DRIVE}/dataset/village_03_orthophoto.tif',
    'village_04': f'{DRIVE}/dataset/village_04_orthophoto.tif',
    'village_05': f'{DRIVE}/dataset/village_05_orthophoto.tif',
    'village_06': f'{DRIVE}/dataset/village_06_orthophoto.tif',
    'village_07': f'{DRIVE}/dataset/village_07_orthophoto.tif',
    'village_08': f'{DRIVE}/dataset/village_08_orthophoto.tif',
    'village_09': f'{DRIVE}/dataset/village_09_orthophoto.tif',
    'village_10': f'{DRIVE}/dataset/village_10_orthophoto.tif',
}

print("Copying villages to Colab SSD...")
for vid, src in sources.items():
    dst = f'{VILLAGES}/{vid}/orthophoto.tif'
    if os.path.exists(dst):
        size = os.path.getsize(dst)/1e9
        print(f"  {vid}: already done ({size:.2f} GB)")
        continue
    if not os.path.exists(src):
        print(f"  {vid}: NOT FOUND — {src}")
        continue
    size = os.path.getsize(src)/1e9
    print(f"  Copying {vid} ({size:.2f} GB)...")
    shutil.copy2(src, dst)
    print(f"    Done")

# Extract SHP files
print("\nExtracting SHP files...")
with zipfile.ZipFile(f'{DRIVE}/dataset/shp_cg.zip','r') as z:
    z.extractall(SHP_CG)
print(f"  CG SHP: {[f for f in os.listdir(SHP_CG) if f.endswith('.shp')]}")

with zipfile.ZipFile(f'{DRIVE}/dataset/shp_pb.zip','r') as z:
    z.extractall(SHP_PB)
print(f"  PB SHP: {[f for f in os.listdir(SHP_PB) if f.endswith('.shp')]}")

# Final check
print("\n" + "="*45)
print("FINAL STATUS")
print("="*45)
ready = 0
for v in sorted(os.listdir(VILLAGES)):
    ortho = f'{VILLAGES}/{v}/orthophoto.tif'
    if os.path.exists(ortho):
        size = os.path.getsize(ortho)/1e9
        print(f"  {v}: READY ({size:.2f} GB)")
        ready += 1
    else:
        print(f"  {v}: MISSING")
print(f"\n  {ready}/10 villages ready")
print(f"  CG SHP: {len([f for f in os.listdir(SHP_CG) if f.endswith('.shp')])} files")
print(f"  PB SHP: {len([f for f in os.listdir(SHP_PB) if f.endswith('.shp')])} files")
if ready == 10:
    print("\n  ALL READY — starting pipeline!")
```

```python
import geopandas as gpd
import os

SHP_CG = '/content/svamitva/data/train/shp-file_cg'
SHP_PB = '/content/svamitva/data/train/shp-file'

for shp_dir, state in [(SHP_CG, 'CG'), (SHP_PB, 'PB')]:
    print(f"\n{'='*55}")
    print(f"STATE: {state}  →  {shp_dir}")
    print('='*55)

    shps = [f for f in os.listdir(shp_dir) if f.endswith('.shp')]

    for shp_name in sorted(shps):
        path = os.path.join(shp_dir, shp_name)
        try:
            gdf = gpd.read_file(path)
            print(f"\n  FILE: {shp_name}")
            print(f"  Rows     : {len(gdf)}")
            print(f"  Geometry : {gdf.geometry.geom_type.unique().tolist()}")
            print(f"  CRS      : {gdf.crs}")
            print(f"  Columns  : {gdf.columns.tolist()}")
            # Show unique values of key columns
            for col in gdf.columns:
                if col.lower() in ['type','class','category','roof','roof_type',
                                   'feature','landuse','built_up','asset']:
                    print(f"  {col} values: {gdf[col].unique().tolist()[:10]}")
            # Show first row
            print(f"  Sample   : {gdf.iloc[0].to_dict() if len(gdf)>0 else 'empty'}")
        except Exception as e:
            print(f"  ERROR reading {shp_name}: {e}")
```

```python
# ═══════════════════════════════════════════════════════
# CELL 9 — Write prepare_tiles.py
# ═══════════════════════════════════════════════════════
code = r'''
import argparse
from pathlib import Path
import numpy as np
import rasterio
from rasterio.windows import Window
from config import (TRAIN_DIR,TRAIN_VILLAGES,VAL_VILLAGES,
                    TILE_SIZE,STEP_SIZE,MIN_VALID_FRACTION,MIN_TILE_BYTES)

def write_tile(data, profile, path):
    p=profile.copy()
    p.update(height=data.shape[-2],width=data.shape[-1],
             count=data.shape[0] if data.ndim==3 else 1,
             compress="lzw",driver="GTiff")
    with rasterio.open(path,"w",**p) as dst:
        dst.write(data) if data.ndim==3 else dst.write(data,1)

def process_village(vdir):
    ortho=vdir/"orthophoto.tif"; mask=vdir/"masks"/"mask.tif"
    if not ortho.exists(): return 0,0
    pd_=vdir/"patches";       pd_.mkdir(exist_ok=True)
    md_=vdir/"mask_patches";  md_.mkdir(exist_ok=True)
    has_mask=mask.exists()
    saved=skipped=0
    with rasterio.open(ortho) as src:
        H,W=src.height,src.width
        profile=src.profile.copy()
        msrc=rasterio.open(mask) if has_mask else None
        mprofile=msrc.profile.copy() if msrc else None
        for r in range(0,H-TILE_SIZE+1,STEP_SIZE):
            for c in range(0,W-TILE_SIZE+1,STEP_SIZE):
                win=Window(c,r,TILE_SIZE,TILE_SIZE)
                img=src.read(window=win)
                rgb=img[:3] if img.shape[0]>=3 else img
                if float(rgb.mean())<5.0:    skipped+=1; continue
                if float(rgb.mean())>245.0:  skipped+=1; continue
                if np.any(rgb>10,axis=0).mean()<MIN_VALID_FRACTION:
                    skipped+=1; continue
                fname=f"patch_{r:05d}_{c:05d}.tif"
                out=pd_/fname
                write_tile(img,profile,out)
                if out.stat().st_size<MIN_TILE_BYTES:
                    out.unlink(); skipped+=1; continue
                if has_mask:
                    msk=msrc.read(1,window=win)
                    write_tile(msk,mprofile,md_/fname)
                saved+=1
        if msrc: msrc.close()
    # Sync
    if has_mask:
        imgs={p.name for p in pd_.glob("*.tif")}
        msks={p.name for p in md_.glob("*.tif")}
        for n in imgs-msks: (pd_/n).unlink()
        for n in msks-imgs: (md_/n).unlink()
    return saved,skipped

def main():
    ap=argparse.ArgumentParser()
    ap.add_argument("--villages",nargs="*",default=TRAIN_VILLAGES+VAL_VILLAGES)
    args=ap.parse_args()
    tot_s=tot_k=0
    for vid in args.villages:
        vdir=TRAIN_DIR/vid
        if not vdir.exists(): continue
        print(f"\n── {vid} ──")
        s,k=process_village(vdir)
        print(f"  saved={s}  skipped={k}")
        tot_s+=s; tot_k+=k
    print(f"\n[done] total saved={tot_s} skipped={tot_k}")

if __name__=="__main__":
    main()
'''
with open('prepare_tiles.py','w') as f:
    f.write(code)
print('prepare_tiles.py written')
```

```python
# ═══════════════════════════════════════════════════════
# CELL 10 — Write model, losses, metrics, dataset
# ═══════════════════════════════════════════════════════

# ── models/svamitva_model.py ──────────────────────────
model_code = r'''
import torch, torch.nn as nn, torch.nn.functional as F
from transformers import SegformerModel, SegformerConfig
from config import BACKBONE,FPN_OUT_CH,ASPP_OUT_CH,NUM_SEG_CLASSES,NUM_ROOF_CLS,NUM_ASSET_CLS

class ASPP(nn.Module):
    def __init__(self,in_ch,out_ch):
        super().__init__()
        self.branches=nn.ModuleList([
            nn.Sequential(nn.Conv2d(in_ch,out_ch,3,padding=d,dilation=d,bias=False),
                          nn.BatchNorm2d(out_ch),nn.ReLU(True))
            for d in [1,6,12,18]])
        self.gap=nn.Sequential(nn.AdaptiveAvgPool2d(1),
                               nn.Conv2d(in_ch,out_ch,1,bias=False),
                               nn.BatchNorm2d(out_ch),nn.ReLU(True))
        self.proj=nn.Sequential(nn.Conv2d(out_ch*5,out_ch,1,bias=False),
                                nn.BatchNorm2d(out_ch),nn.ReLU(True),nn.Dropout2d(0.1))
    def forward(self,x):
        h,w=x.shape[2:]
        feats=[b(x) for b in self.branches]
        feats.append(F.interpolate(self.gap(x),(h,w),mode="bilinear",align_corners=False))
        return self.proj(torch.cat(feats,1))

class FPN(nn.Module):
    CH=[64,128,320,512]
    def __init__(self,out_ch):
        super().__init__()
        self.lat=[nn.Conv2d(c,out_ch,1,bias=False) for c in self.CH]
        self.lat=nn.ModuleList(self.lat)
        self.smooth=nn.ModuleList([
            nn.Sequential(nn.Conv2d(out_ch,out_ch,3,padding=1,bias=False),
                          nn.BatchNorm2d(out_ch),nn.ReLU(True))
            for _ in self.CH])
    def forward(self,feats):
        lats=[l(f) for l,f in zip(self.lat,feats)]
        for i in range(len(lats)-2,-1,-1):
            lats[i]+=F.interpolate(lats[i+1],lats[i].shape[2:],mode="bilinear",align_corners=False)
        t=lats[0].shape[2:]
        outs=[F.interpolate(s(l),t,mode="bilinear",align_corners=False) if s(l).shape[2:]!=t else s(l)
              for s,l in zip(self.smooth,lats)]
        return torch.stack(outs).sum(0)

class ClassHead(nn.Module):
    def __init__(self,in_ch,n):
        super().__init__()
        self.pool=nn.AdaptiveAvgPool2d(1)
        self.mlp=nn.Sequential(nn.Flatten(),nn.Linear(in_ch,128),
                                nn.ReLU(True),nn.Dropout(0.3),nn.Linear(128,n))
    def forward(self,x): return self.mlp(self.pool(x))

class DetHead(nn.Module):
    def __init__(self,in_ch,nc=2):
        super().__init__()
        def blk(out): return nn.Sequential(
            nn.Conv2d(in_ch,in_ch,3,padding=1,bias=False),
            nn.BatchNorm2d(in_ch),nn.ReLU(True),nn.Conv2d(in_ch,out,1))
        self.cls=blk(nc); self.reg=blk(4); self.ctr=blk(1)
    def forward(self,x): return self.cls(x),torch.exp(self.reg(x)),self.ctr(x)

class SVAMITVAModel(nn.Module):
    def __init__(self):
        super().__init__()
        cfg=SegformerConfig.from_pretrained(BACKBONE)
        self.backbone=SegformerModel.from_pretrained(BACKBONE,config=cfg,ignore_mismatched_sizes=True)
        self.fpn=FPN(FPN_OUT_CH)
        self.aspp=ASPP(FPN_OUT_CH,ASPP_OUT_CH)
        self.seg_head=nn.Sequential(
            nn.Conv2d(ASPP_OUT_CH,128,3,padding=1,bias=False),
            nn.BatchNorm2d(128),nn.ReLU(True),nn.Conv2d(128,NUM_SEG_CLASSES,1))
        self.det_head=DetHead(ASPP_OUT_CH)
        self.roof_head=ClassHead(ASPP_OUT_CH,NUM_ROOF_CLS)
        self.asset_head=ClassHead(ASPP_OUT_CH,NUM_ASSET_CLS)

    def _feats(self,x):
        out=self.backbone(pixel_values=x,output_hidden_states=True)
        h=out.hidden_states
        return list(h[-4:]) if len(h)>=4 else list(h)

    def forward(self,x,stage=2):
        h,w=x.shape[2:]
        ctx=self.aspp(self.fpn(self._feats(x)))
        seg=F.interpolate(self.seg_head(ctx),(h,w),mode="bilinear",align_corners=False)
        dc,dr,dct=self.det_head(ctx)
        out=dict(seg=seg,det_cls=dc,det_reg=dr,det_ctr=dct)
        if stage==2:
            out["roof"]=self.roof_head(ctx)
            out["asset"]=self.asset_head(ctx)
        return out

    def freeze_cls_heads(self):
        for p in list(self.roof_head.parameters())+list(self.asset_head.parameters()):
            p.requires_grad=False
    def unfreeze_all(self):
        for p in self.parameters(): p.requires_grad=True

def build_model(): return SVAMITVAModel()
'''
with open('models/svamitva_model.py','w') as f:
    f.write(model_code)

# ── utils/losses.py ──────────────────────────────────
losses_code = r'''
import torch, torch.nn as nn, torch.nn.functional as F
from config import CLASS_WEIGHTS, DEVICE

class DiceLoss(nn.Module):
    def __init__(self,nc=10,smooth=1.0):
        super().__init__(); self.nc=nc; self.smooth=smooth
    def forward(self,logits,targets):
        p=F.softmax(logits,1)
        oh=F.one_hot(targets,self.nc).permute(0,3,1,2).float()
        inter=(p*oh).sum((0,2,3))
        union=p.sum((0,2,3))+oh.sum((0,2,3))
        return 1.0-((2*inter+self.smooth)/(union+self.smooth)).mean()

class SegLoss(nn.Module):
    def __init__(self):
        super().__init__()
        self.ce=nn.CrossEntropyLoss(weight=CLASS_WEIGHTS.to(DEVICE),ignore_index=255)
        self.dice=DiceLoss()
    def forward(self,logits,targets):
        return 0.5*self.ce(logits,targets)+0.5*self.dice(logits,targets)

def focal_loss(pred,tgt,g=2.0,a=0.25):
    p=torch.sigmoid(pred)
    ce=F.binary_cross_entropy_with_logits(pred,tgt,reduction="none")
    pt=p*tgt+(1-p)*(1-tgt)
    return (a*((1-pt)**g)*ce).mean()

class DetLoss(nn.Module):
    def forward(self,dc,dr,dct,tc,tr_,tctr):
        pos=tc.sum(1)>0
        cl=focal_loss(dc,tc.float())
        rl=dr.sum()*0.0
        ctl=F.binary_cross_entropy_with_logits(
            dct.squeeze(1)[pos],tctr[pos].float()
        ) if pos.any() else dct.sum()*0.0
        return cl+rl+ctl

roof_loss=nn.CrossEntropyLoss()
asset_loss=nn.CrossEntropyLoss()
'''
with open('utils/losses.py','w') as f:
    f.write(losses_code)

# ── utils/metrics.py ─────────────────────────────────
metrics_code = r'''
import numpy as np
from config import NUM_SEG_CLASSES, CLASS_NAMES, BUILDING_CLASSES, ASSET_CLASSES

class SegMetrics:
    def __init__(self,nc=NUM_SEG_CLASSES):
        self.nc=nc; self.mat=np.zeros((nc,nc),dtype=np.int64)
    def update(self,pred,target):
        p=pred.cpu().numpy().ravel(); t=target.cpu().numpy().ravel()
        v=(t>=0)&(t<self.nc)
        self.mat+=np.bincount(self.nc*t[v]+p[v],minlength=self.nc**2).reshape(self.nc,self.nc)
    def iou_per_class(self):
        d=np.diag(self.mat); u=self.mat.sum(1)+self.mat.sum(0)-d
        return np.where(u>0,d/u,np.nan)
    def miou(self): return float(np.nanmean(self.iou_per_class()))
    def pixel_accuracy(self):
        t=self.mat.sum(); return float(np.diag(self.mat).sum()/t) if t>0 else 0.0
    def building_iou(self):
        iou=self.iou_per_class()
        vals=[iou[i] for i in BUILDING_CLASSES if not np.isnan(iou[i])]
        return float(np.mean(vals)) if vals else 0.0
    def asset_iou(self):
        iou=self.iou_per_class()
        vals=[iou[i] for i in ASSET_CLASSES if not np.isnan(iou[i])]
        return float(np.mean(vals)) if vals else 0.0
    def report(self):
        iou=self.iou_per_class()
        r={f"iou_{n}":float(iou[i]) if not np.isnan(iou[i]) else 0.0
           for i,n in enumerate(CLASS_NAMES)}
        r["mIoU"]=self.miou()
        r["pixel_acc"]=self.pixel_accuracy()
        r["building_iou"]=self.building_iou()
        r["asset_iou"]=self.asset_iou()
        r["target_95pct_met"]=r["pixel_acc"]>=0.95
        return r
    def reset(self): self.mat[:]=0
'''
with open('utils/metrics.py','w') as f:
    f.write(metrics_code)

# ── utils/dataset.py ─────────────────────────────────
dataset_code = r'''
from pathlib import Path
import numpy as np, rasterio, torch
from torch.utils.data import Dataset
import albumentations as A
from albumentations.pytorch import ToTensorV2
from config import TRAIN_DIR, TRAIN_VILLAGES, VAL_VILLAGES

MEAN=(0.485,0.456,0.406); STD=(0.229,0.224,0.225)

def train_aug():
    return A.Compose([
        A.HorizontalFlip(p=0.5), A.VerticalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.ShiftScaleRotate(shift_limit=0.05,scale_limit=0.15,
                           rotate_limit=20,border_mode=0,p=0.6),
        A.ColorJitter(brightness=0.25,contrast=0.25,saturation=0.15,hue=0.08,p=0.6),
        A.CLAHE(clip_limit=4.0,tile_grid_size=(8,8),p=0.4),
        A.OneOf([A.GaussianBlur(blur_limit=(3,7),p=1.0),
                 A.GaussNoise(var_limit=(10,60),p=1.0)],p=0.35),
        A.CoarseDropout(max_holes=8,max_height=32,max_width=32,
                        min_holes=1,fill_value=0,p=0.2),
        A.Normalize(mean=MEAN,std=STD), ToTensorV2(),
    ])

def val_aug():
    return A.Compose([A.Normalize(mean=MEAN,std=STD), ToTensorV2()])

def read_rgb(path):
    with rasterio.open(path) as src: img=src.read()
    return np.transpose(img[:3].astype(np.uint8),(1,2,0))

def read_mask(path):
    with rasterio.open(path) as src: return src.read(1).astype(np.int64)

class TileDataset(Dataset):
    def __init__(self,villages,augment=True):
        self.aug=train_aug() if augment else val_aug()
        self.items=[]
        for vid in villages:
            pd_=TRAIN_DIR/vid/"patches"
            md_=TRAIN_DIR/vid/"mask_patches"
            if not pd_.exists(): continue
            for p in sorted(pd_.glob("*.tif")):
                m=md_/p.name
                if m.exists(): self.items.append((p,m))
    def __len__(self): return len(self.items)
    def __getitem__(self,i):
        ip,mp=self.items[i]
        img=read_rgb(ip); mask=read_mask(mp)
        a=self.aug(image=img,mask=mask)
        return a["image"],torch.as_tensor(a["mask"],dtype=torch.long)
'''
with open('utils/dataset.py','w') as f:
    f.write(dataset_code)

print('All code files written:')
import os
for f in ['models/svamitva_model.py','utils/losses.py','utils/metrics.py','utils/dataset.py']:
    print(f'  {f}: {os.path.getsize(f)} bytes')
```

```python
# ═══════════════════════════════════════════════════════
# CELL 11 — Write train.py
# ═══════════════════════════════════════════════════════
train_code = r'''
import argparse, json, time
from pathlib import Path
import torch, torch.nn as nn
from torch.utils.data import DataLoader
from torch.cuda.amp import GradScaler, autocast
from config import *
from models.svamitva_model import build_model
from utils.dataset import TileDataset
from utils.losses import SegLoss, DetLoss
from utils.metrics import SegMetrics

def make_det_tgt(masks, device):
    B=masks.shape[0]; H=W=masks.shape[-1]//4
    tc=torch.zeros(B,2,H,W,device=device)
    tr_=torch.ones(B,4,H,W,device=device)
    ms=torch.nn.functional.interpolate(
        masks.unsqueeze(1).float(),(H,W),mode="nearest").squeeze(1).long()
    # buildings=1-4, assets=7-9
    tc[:,0]=(torch.isin(ms,torch.tensor([1,2,3,4],device=device))).float()
    tc[:,1]=(torch.isin(ms,torch.tensor([7,8,9],device=device))).float()
    tctr=tc.max(1).values
    return tc,tr_,tctr

def warmup_cosine(opt, epoch, total, warmup, base_lr):
    import math
    lr=(base_lr*(epoch+1)/warmup) if epoch<warmup else (
       base_lr*0.5*(1+math.cos(math.pi*(epoch-warmup)/max(1,total-warmup))))
    for pg in opt.param_groups: pg["lr"]=lr
    return lr

def train_epoch(model, loader, opt, scaler, sc, dc, stage, device, ws, wd):
    model.train(); total=0.0
    for imgs,masks in loader:
        imgs=imgs.to(device); masks=masks.to(device)
        tc,tr_,tctr=make_det_tgt(masks,device)
        with autocast(enabled=AMP):
            out=model(imgs,stage=stage)
            loss=ws*sc(out["seg"],masks)+wd*dc(out["det_cls"],out["det_reg"],
                 out["det_ctr"],tc,tr_,tctr)
        opt.zero_grad(set_to_none=True)
        scaler.scale(loss).backward()
        scaler.unscale_(opt)
        nn.utils.clip_grad_norm_(model.parameters(),GRAD_CLIP)
        scaler.step(opt); scaler.update()
        total+=loss.item()
    return total/len(loader)

@torch.no_grad()
def validate(model, loader, sc, stage, device):
    model.eval(); metrics=SegMetrics(); vl=0.0
    for imgs,masks in loader:
        imgs=imgs.to(device); masks=masks.to(device)
        with autocast(enabled=AMP):
            out=model(imgs,stage=stage)
            vl+=sc(out["seg"],masks).item()
        metrics.update(out["seg"].argmax(1),masks)
    r=metrics.report(); r["val_loss"]=vl/len(loader)
    return r

def run_stage(model,stage,epochs,lr,train_ds,val_ds,device,history,start=0):
    tl=DataLoader(train_ds,BATCH_SIZE,shuffle=True,num_workers=NUM_WORKERS,
                  pin_memory=PIN_MEMORY,drop_last=True)
    vl=DataLoader(val_ds,BATCH_SIZE,shuffle=False,num_workers=NUM_WORKERS,pin_memory=PIN_MEMORY)
    opt=torch.optim.AdamW(filter(lambda p:p.requires_grad,model.parameters()),
                          lr=lr,weight_decay=WEIGHT_DECAY)
    scaler=GradScaler(enabled=AMP)
    sc=SegLoss().to(device); dc=DetLoss().to(device)
    ws,wd=(S1_W_SEG,S1_W_DET) if stage==1 else (S2_W_SEG,S2_W_DET)
    best=float("inf")
    for epoch in range(start,epochs):
        lr_=warmup_cosine(opt,epoch,epochs,WARMUP_EPOCHS,lr)
        t0=time.time()
        tl_=train_epoch(model,tl,opt,scaler,sc,dc,stage,device,ws,wd)
        vr=validate(model,vl,sc,stage,device)
        elapsed=time.time()-t0
        row=dict(stage=stage,epoch=epoch+1,lr=round(lr_,7),
                 train_loss=round(tl_,4),
                 **{k:round(v,4) for k,v in vr.items()},
                 time_s=round(elapsed,1))
        history.append(row)
        status="PASS" if vr["target_95pct_met"] else "----"
        print(f"[S{stage} E{epoch+1:03d}/{epochs}] "
              f"lr={lr_:.2e} train={tl_:.4f} "
              f"val={vr[chr(39)+'val_loss'+chr(39)]:.4f} "
              f"mIoU={vr[chr(39)+'mIoU'+chr(39)]:.4f} "
              f"pix={vr[chr(39)+'pixel_acc'+chr(39)]:.4f} "
              f"bldg={vr[chr(39)+'building_iou'+chr(39)]:.4f} "
              f"asset={vr[chr(39)+'asset_iou'+chr(39)]:.4f} "
              f"{status} ({elapsed:.0f}s)")
        torch.save(dict(epoch=epoch+1,stage=stage,
                        state_dict=model.state_dict(),history=history),
                   CKPT_DIR/"model_last.pt")
        if vr["val_loss"]<best:
            best=vr["val_loss"]
            torch.save(dict(epoch=epoch+1,stage=stage,
                            state_dict=model.state_dict(),history=history),
                       CKPT_DIR/"model_best.pt")
            print(f"  *** best val_loss={best:.4f} pix={vr[chr(39)+'pixel_acc'+chr(39)]:.4f} ***")
    torch.save(dict(state_dict=model.state_dict(),history=history),
               CKPT_DIR/f"model_stage{stage}.pt")
    print(f"[Stage {stage} done] best_val_loss={best:.4f}")

def main():
    ap=argparse.ArgumentParser()
    ap.add_argument("--stage",type=int,default=1)
    ap.add_argument("--resume",type=str,default=None)
    args=ap.parse_args()
    print(f"[train] device={DEVICE} AMP={AMP}")
    model=build_model().to(DEVICE)
    history=[]; start=0
    if args.resume and Path(args.resume).exists():
        ck=torch.load(args.resume,map_location=DEVICE)
        model.load_state_dict(ck["state_dict"])
        history=ck.get("history",[]); start=ck.get("epoch",0)
        print(f"  resumed epoch {start}")
    train_ds=TileDataset(TRAIN_VILLAGES,True)
    val_ds=TileDataset(VAL_VILLAGES,False)
    print(f"  train={len(train_ds)} val={len(val_ds)}")
    if args.stage==1:
        model.freeze_cls_heads()
        print("=== STAGE 1 ===")
        run_stage(model,1,STAGE1_EPOCHS,STAGE1_LR,train_ds,val_ds,DEVICE,history,start)
    else:
        model.unfreeze_all()
        print("=== STAGE 2 ===")
        run_stage(model,2,STAGE2_EPOCHS,STAGE2_LR,train_ds,val_ds,DEVICE,history,start)
    with open(OUTPUT_DIR/"training_history.json","w") as f:
        json.dump(history,f,indent=2)
    print("[done] training complete")

if __name__=="__main__": main()
'''
with open('train.py','w') as f:
    f.write(train_code)
print('train.py written')
```

```python
# Fix — call main() directly without sys.argv
import os, sys
os.chdir('/content/svamitva')
sys.path.insert(0, '/content/svamitva')

# Reload all modules fresh
for mod in list(sys.modules.keys()):
    if any(x in mod for x in ['config','prepare','models','utils']):
        del sys.modules[mod]

# Import and run directly
from config import TRAIN_DIR, SHP_CG_DIR, SHP_PB_DIR, PUNJAB_IDS
from pathlib import Path
import numpy as np
import rasterio
from rasterio.features import rasterize
from shapely.geometry import mapping
import geopandas as gpd
import csv, warnings
warnings.filterwarnings("ignore")

# Paste the functions from prepare_data.py
BG=0; BLDG_RCC=1; BLDG_TILED=2; BLDG_TIN=3; BLDG_OTHER=4
ROAD=5; WATER=6; ASSET_DT=7; ASSET_OHT=8; ASSET_WELL=9

ROOF_MAP  = {1:BLDG_RCC,2:BLDG_TILED,3:BLDG_TIN,4:BLDG_OTHER}
ASSET_MAP = {1:ASSET_DT,2:ASSET_OHT,3:ASSET_WELL,4:ASSET_WELL,
             5:ASSET_WELL,6:ASSET_OHT,7:ASSET_DT,8:ASSET_DT,
             9:ASSET_WELL,10:ASSET_DT,11:ASSET_DT,12:ASSET_OHT,
             13:ASSET_OHT,14:ASSET_WELL}
RMAP = {1:"RCC",2:"Tiled",3:"Tin",4:"Others"}
UMAP = {1:"DT",2:"OHT",3:"Well",4:"Handpump",5:"Tubewell",
        6:"Tank",7:"Solar",8:"Light",9:"Borewell",
        10:"DT",11:"DT",12:"OHT",13:"OHT",14:"Well"}

def get_shp_dir(vid):
    v = vid.replace("village_","").zfill(2)
    return SHP_PB_DIR if v in PUNJAB_IDS else SHP_CG_DIR

def load_shp(shp_dir, name, crs):
    for n in [name, name+"_type", name+"_Poly_", name+"_Poly"]:
        p = shp_dir / f"{n}.shp"
        if p.exists():
            try:
                gdf = gpd.read_file(p)
                if gdf is None or len(gdf)==0: continue
                if gdf.crs is None: gdf = gdf.set_crs("EPSG:3857")
                return gdf.to_crs(crs)
            except Exception as e:
                print(f"    CRS warn {n}: {e}")
    return None

def burn(pairs, transform, shape):
    valid = [(mapping(g),v) for g,v in pairs
             if g is not None and not g.is_empty]
    if not valid: return np.zeros(shape, dtype=np.uint8)
    return rasterize(valid, out_shape=shape, transform=transform,
                     fill=0, dtype=np.uint8, all_touched=True)

def process_village(vdir, shp_dir):
    ortho = vdir/"orthophoto.tif"
    if not ortho.exists(): print("  [skip] no orthophoto"); return
    (vdir/"masks").mkdir(exist_ok=True)
    with rasterio.open(ortho) as src:
        meta=src.meta.copy(); H,W=src.height,src.width
        crs=src.crs; tr=src.transform
    print(f"  CRS={crs}  {W}x{H}")
    meta.update(count=1,dtype="uint8",compress="lzw",nodata=None)
    shape=(H,W); mask=np.zeros(shape,dtype=np.uint8)

    w=load_shp(shp_dir,"Water_Body",crs)
    if w is not None and len(w):
        l=burn([(g,WATER) for g in w.geometry],tr,shape)
        mask[l>0]=WATER; print(f"  water_poly    : {len(w)}")

    wl=load_shp(shp_dir,"Water_Body_Line",crs)
    if wl is not None and len(wl):
        polys=[g.buffer(3.0) for g in wl.geometry if g and not g.is_empty]
        l=burn([(g,WATER) for g in polys],tr,shape)
        mask[l>0]=WATER; print(f"  water_lines   : {len(wl)}")

    r=load_shp(shp_dir,"Road",crs)
    if r is not None and len(r):
        l=burn([(g,ROAD) for g in r.geometry],tr,shape)
        mask[l>0]=ROAD; print(f"  road_poly     : {len(r)}")

    rc=load_shp(shp_dir,"Road_Centre_Line",crs)
    if rc is not None and len(rc):
        polys=[]
        for _,row in rc.iterrows():
            if row.geometry is None or row.geometry.is_empty: continue
            w_=float(row.get("Width",4.0) or 4.0)
            polys.append(row.geometry.buffer(max(2.0,min(w_,20.0))/2))
        l=burn([(g,ROAD) for g in polys],tr,shape)
        mask[l>0]=ROAD; print(f"  road_cl       : {len(rc)}")

    b=load_shp(shp_dir,"Built_Up_Area_type",crs)
    if b is None: b=load_shp(shp_dir,"Built_Up_Area_typ",crs)
    if b is not None and len(b):
        pairs=[]
        for _,row in b.iterrows():
            if row.geometry is None or row.geometry.is_empty: continue
            rt=int(row.get("Roof_type",4) or 4)
            pairs.append((row.geometry, ROOF_MAP.get(rt,BLDG_OTHER)))
        l=burn(pairs,tr,shape)
        mask[l>0]=l[l>0]
        print(f"  buildings     : {len(b)}")
        for rv,cn in [(1,"RCC"),(2,"Tiled"),(3,"Tin"),(4,"Other")]:
            cnt=(b["Roof_type"]==rv).sum()
            if cnt>0: print(f"    {cn}: {cnt}")

    u=None
    for nm in ["Utility_Poly_","Utility_Poly","Utility"]:
        u=load_shp(shp_dir,nm,crs)
        if u is not None and len(u): break
    if u is not None and len(u):
        pairs=[]
        for _,row in u.iterrows():
            if row.geometry is None or row.geometry.is_empty: continue
            ut=int(row.get("Utility_Ty",1) or 1)
            g=row.geometry
            if g.geom_type=="Point": g=g.buffer(5)
            pairs.append((g, ASSET_MAP.get(ut,ASSET_DT)))
        l=burn(pairs,tr,shape)
        mask[l>0]=l[l>0]
        print(f"  utilities     : {len(u)}")

    with rasterio.open(vdir/"masks"/"mask.tif","w",**meta) as dst:
        dst.write(mask,1)
    unique,counts=np.unique(mask,return_counts=True)
    total=mask.size
    cls_n=["bg","RCC","Tiled","Tin","Other","road","water","DT","OHT","Well"]
    print(f"  non-bg: {100*(1-counts[0]/total):.2f}%")
    for u_,c_ in zip(unique,counts):
        if u_>0: print(f"    {cls_n[u_]:12}: {c_:>10,} px")

# Run all 10 villages
ALL_VILLAGES = [
    "village_01","village_02","village_03","village_04","village_05",
    "village_06","village_07","village_08","village_09","village_10"
]

for vid in ALL_VILLAGES:
    vdir = TRAIN_DIR / vid
    if not vdir.exists():
        print(f"[skip] {vid}")
        continue
    shp_dir = get_shp_dir(vid)
    print(f"\n{'='*50}")
    print(f"Village: {vid}  SHP: {shp_dir.name}")
    try:
        process_village(vdir, shp_dir)
    except Exception as e:
        print(f"  ERROR: {e}")
        import traceback; traceback.print_exc()

print("\n[done] all villages processed")
```

```python
# Fix village_04 CRS using rasterio only — no gdalwarp needed
import rasterio
from rasterio.crs import CRS
from rasterio.warp import calculate_default_transform, reproject, Resampling
import numpy as np
import os

ortho = '/content/svamitva/data/train_villages/village_04/orthophoto.tif'
fixed = '/content/svamitva/data/train_villages/village_04/orthophoto_fixed.tif'

print("Reading village_04...")
with rasterio.open(ortho) as src:
    print(f"  Current CRS  : {src.crs}")
    print(f"  Current size : {src.width}x{src.height}")
    print(f"  Bounds       : {src.bounds}")

    src_crs  = src.crs
    dst_crs  = CRS.from_epsg(32643)   # correct UTM zone for CG India

    transform, width, height = calculate_default_transform(
        src_crs, dst_crs, src.width, src.height, *src.bounds
    )

    meta = src.meta.copy()
    meta.update(crs=dst_crs, transform=transform,
                width=width, height=height,
                compress='lzw')

    print(f"\nReprojecting to EPSG:32643...")
    print(f"  New size: {width}x{height}")

    with rasterio.open(fixed, 'w', **meta) as dst:
        for i in range(1, src.count+1):
            reproject(
                source=rasterio.band(src, i),
                destination=rasterio.band(dst, i),
                src_transform=src.transform,
                src_crs=src_crs,
                dst_transform=transform,
                dst_crs=dst_crs,
                resampling=Resampling.bilinear
            )

os.replace(fixed, ortho)
print("Done — village_04 fixed!")

with rasterio.open(ortho) as src:
    print(f"  Final CRS  : {src.crs}")
    print(f"  Final size : {src.width}x{src.height}")
```

```python
import os, sys, warnings, csv
import numpy as np
import rasterio
from rasterio.features import rasterize
from rasterio.crs import CRS
from shapely.geometry import mapping
import geopandas as gpd
from pathlib import Path
import glob
warnings.filterwarnings("ignore")

os.chdir('/content/svamitva')
sys.path.insert(0, '/content/svamitva')
for mod in list(sys.modules.keys()):
    if 'config' in mod: del sys.modules[mod]
from config import TRAIN_DIR, SHP_CG_DIR, SHP_PB_DIR

VILLAGE_MAP = {
    'village_01': ('PB', 'Nadala'),
    'village_02': ('PB', 'fattu_bhila'),
    'village_03': ('PB', 'Bagga'),
    'village_04': ('CG', 'KUTRU'),
    'village_05': ('CG', 'AWAPALLI'),
    'village_06': ('CG', 'JHODIYAWADAM'),
    'village_07': ('CG', 'badetumnar'),
    'village_08': ('PB', 'Tugalwala'),
    'village_09': ('CG', 'GHOTPAL'),
    'village_10': ('PB', 'timmowal'),
}

BG=0;BLDG_RCC=1;BLDG_TILED=2;BLDG_TIN=3;BLDG_OTHER=4
ROAD=5;WATER=6;ASSET_DT=7;ASSET_OHT=8;ASSET_WELL=9
ROOF_MAP={1:BLDG_RCC,2:BLDG_TILED,3:BLDG_TIN,4:BLDG_OTHER}
ASSET_MAP={1:ASSET_DT,2:ASSET_OHT,3:ASSET_WELL,4:ASSET_WELL,
           5:ASSET_WELL,6:ASSET_OHT,7:ASSET_DT,8:ASSET_DT,
           9:ASSET_WELL,10:ASSET_DT,11:ASSET_DT,12:ASSET_OHT,
           13:ASSET_OHT,14:ASSET_WELL}
RMAP={1:"RCC",2:"Tiled",3:"Tin",4:"Others"}
UMAP={1:"DT",2:"OHT",3:"Well",4:"Handpump",5:"Tubewell",
      6:"Tank",7:"Solar",8:"Light",9:"Borewell",
      10:"DT",11:"DT",12:"OHT",13:"OHT",14:"Well"}

def filt(gdf, vname):
    if gdf is None or len(gdf)==0: return None
    if 'Village_Na' not in gdf.columns: return gdf
    out = gdf[gdf['Village_Na'] == vname]
    return out if len(out)>0 else None

def get_utility(layers, vname):
    for key in ['Utility_Poly_','Utility_Poly','Utility']:
        if key in layers:
            result = filt(layers[key], vname)
            if result is not None and len(result)>0:
                return result
    return None

def burn(pairs, transform, shape):
    valid=[(mapping(g),v) for g,v in pairs
           if g is not None and not g.is_empty]
    if not valid: return np.zeros(shape,dtype=np.uint8)
    return rasterize(valid,out_shape=shape,transform=transform,
                     fill=0,dtype=np.uint8,all_touched=True)

def load_all(shp_dir, names):
    out={}
    for n in names:
        p=shp_dir/f'{n}.shp'
        if p.exists():
            out[n]=gpd.read_file(str(p))
    return out

print("Loading SHP layers...")
PB = load_all(SHP_PB_DIR, ['Built_Up_Area_typ','Road','Road_Centre_Line',
                             'Water_Body','Water_Body_Line','Utility_Poly_','Utility'])
CG = load_all(SHP_CG_DIR, ['Built_Up_Area_type','Road','Road_Centre_Line',
                             'Water_Body','Water_Body_Line','Utility_Poly_','Utility'])
print(f"PB: {list(PB.keys())}")
print(f"CG: {list(CG.keys())}")

def process(vid):
    state, vname = VILLAGE_MAP[vid]
    layers = PB if state=='PB' else CG
    bkey   = 'Built_Up_Area_typ' if state=='PB' else 'Built_Up_Area_type'

    vdir  = TRAIN_DIR / vid
    ortho = vdir / "orthophoto.tif"
    if not ortho.exists(): print("  no orthophoto"); return
    (vdir/"masks").mkdir(exist_ok=True)

    with rasterio.open(ortho) as src:
        meta=src.meta.copy(); H,W=src.height,src.width
        crs=src.crs; tr=src.transform
    print(f"  CRS={crs}  {W}x{H}")
    meta.update(count=1,dtype="uint8",compress="lzw",nodata=None)
    shape=(H,W); mask=np.zeros(shape,dtype=np.uint8)

    for lname, cls in [('Water_Body',WATER),('Road',ROAD)]:
        gdf=filt(layers.get(lname),vname)
        if gdf is not None and len(gdf)>0:
            try:
                gdf=gdf.to_crs(crs)
                l=burn([(g,cls) for g in gdf.geometry],tr,shape)
                mask[l>0]=cls; print(f"  {lname}: {len(gdf)}")
            except Exception as e: print(f"  {lname} err: {e}")

    wl=filt(layers.get('Water_Body_Line'),vname)
    if wl is not None and len(wl)>0:
        try:
            wl=wl.to_crs(crs)
            polys=[g.buffer(3.0) for g in wl.geometry if g and not g.is_empty]
            l=burn([(g,WATER) for g in polys],tr,shape)
            mask[l>0]=WATER; print(f"  Water_Line: {len(wl)}")
        except Exception as e: print(f"  water_line err: {e}")

    rc=filt(layers.get('Road_Centre_Line'),vname)
    if rc is not None and len(rc)>0:
        try:
            rc=rc.to_crs(crs)
            polys=[]
            for _,row in rc.iterrows():
                if row.geometry is None or row.geometry.is_empty: continue
                w_=float(row.get("Width",4.0) or 4.0)
                polys.append(row.geometry.buffer(max(2.0,min(w_,20.0))/2))
            l=burn([(g,ROAD) for g in polys],tr,shape)
            mask[l>0]=ROAD; print(f"  Road_CL: {len(rc)}")
        except Exception as e: print(f"  road_cl err: {e}")

    b=filt(layers.get(bkey),vname)
    if b is not None and len(b)>0:
        try:
            b=b.to_crs(crs)
            pairs=[]
            for _,row in b.iterrows():
                if row.geometry is None or row.geometry.is_empty: continue
                rt=int(row.get("Roof_type",4) or 4)
                pairs.append((row.geometry,ROOF_MAP.get(rt,BLDG_OTHER)))
            l=burn(pairs,tr,shape)
            mask[l>0]=l[l>0]
            print(f"  Buildings: {len(b)}")
            for rv,cn in [(1,"RCC"),(2,"Tiled"),(3,"Tin"),(4,"Other")]:
                cnt=(b["Roof_type"]==rv).sum()
                if cnt>0: print(f"    {cn}: {cnt}")
            with open(vdir/"roof_labels.csv","w",newline="") as f_:
                w_=csv.writer(f_)
                w_.writerow(["id","roof_type","roof_name","cx","cy","area_m2"])
                for i,row in b.iterrows():
                    c_=row.geometry.centroid
                    rt=int(row.get("Roof_type",4) or 4)
                    w_.writerow([i,rt,RMAP.get(rt,"Others"),
                                 round(c_.x,2),round(c_.y,2),
                                 round(float(row.get("Area_Sqm",0) or 0),2)])
        except Exception as e: print(f"  buildings err: {e}")

    u=get_utility(layers,vname)
    if u is not None and len(u)>0:
        try:
            u=u.to_crs(crs)
            pairs=[]
            for _,row in u.iterrows():
                if row.geometry is None or row.geometry.is_empty: continue
                ut=int(row.get("Utility_Ty",1) or 1)
                g=row.geometry
                if g.geom_type=="Point": g=g.buffer(5)
                pairs.append((g,ASSET_MAP.get(ut,ASSET_DT)))
            l=burn(pairs,tr,shape)
            mask[l>0]=l[l>0]
            print(f"  Utilities: {len(u)}")
            with open(vdir/"asset_labels.csv","w",newline="") as f_:
                w_=csv.writer(f_)
                w_.writerow(["id","utility_ty","name","x","y"])
                for i,row in u.iterrows():
                    c_=row.geometry.centroid
                    ut=int(row.get("Utility_Ty",1) or 1)
                    w_.writerow([i,ut,UMAP.get(ut,"Other"),
                                 round(c_.x,2),round(c_.y,2)])
        except Exception as e: print(f"  utilities err: {e}")

    with rasterio.open(vdir/"masks"/"mask.tif","w",**meta) as dst:
        dst.write(mask,1)
    unique,counts=np.unique(mask,return_counts=True)
    total=mask.size
    cls_n=["bg","RCC","Tiled","Tin","Other","road","water","DT","OHT","Well"]
    print(f"  non-bg: {100*(1-counts[0]/total):.2f}%")
    for u_,c_ in zip(unique,counts):
        if u_>0: print(f"    {cls_n[u_]:10}: {c_:>10,} px")
    print(f"  DONE")

# Check which villages already have masks
done = set()
for m in glob.glob('/content/svamitva/data/train_villages/*/masks/mask.tif'):
    done.add(m.split('/')[5])
print(f"\nAlready done: {sorted(done)}")

# Process only missing villages
all_villages = ['village_01','village_02','village_03','village_04',
                'village_05','village_06','village_07','village_08',
                'village_09','village_10']
todo = [v for v in all_villages if v not in done]
print(f"Need to process: {todo}")

for vid in todo:
    print(f"\n{'='*50}")
    print(f"Village: {vid}  ({VILLAGE_MAP[vid][1]})")
    try:
        process(vid)
    except Exception as e:
        print(f"  ERROR: {e}")
        import traceback; traceback.print_exc()

# Final summary
print(f"\n{'='*50}")
print("FINAL SUMMARY")
masks = glob.glob('/content/svamitva/data/train_villages/*/masks/mask.tif')
print(f"MASKS: {len(masks)}/10")
for m in sorted(masks):
    vid=m.split('/')[5]
    with rasterio.open(m) as s: data=s.read(1)
    u=np.unique(data); nb=round(100*(data>0).sum()/data.size,2)
    print(f"  {vid}: classes={u.tolist()}  non-bg={nb}%")
```

```python
import os, sys, warnings, csv, glob
import numpy as np
import rasterio
from rasterio.features import rasterize
from shapely.geometry import mapping
import geopandas as gpd
from pathlib import Path
warnings.filterwarnings("ignore")

os.chdir('/content/svamitva')
sys.path.insert(0, '/content/svamitva')
for mod in list(sys.modules.keys()):
    if 'config' in mod: del sys.modules[mod]
from config import TRAIN_DIR, SHP_CG_DIR, SHP_PB_DIR

VILLAGE_MAP = {
    'village_01': ('PB', 'Nadala'),
    'village_02': ('PB', 'fattu_bhila'),
    'village_03': ('PB', 'Bagga'),
    'village_04': ('CG', 'KUTRU'),
    'village_05': ('CG', 'AWAPALLI'),
    'village_06': ('CG', 'JHODIYAWADAM'),
    'village_07': ('CG', 'badetumnar'),
    'village_08': ('PB', 'Tugalwala'),
    'village_09': ('CG', 'GHOTPAL'),
    'village_10': ('PB', 'timmowal'),
}

BG=0;BLDG_RCC=1;BLDG_TILED=2;BLDG_TIN=3;BLDG_OTHER=4
ROAD=5;WATER=6;ASSET_DT=7;ASSET_OHT=8;ASSET_WELL=9
ROOF_MAP={1:BLDG_RCC,2:BLDG_TILED,3:BLDG_TIN,4:BLDG_OTHER}
ASSET_MAP={1:ASSET_DT,2:ASSET_OHT,3:ASSET_WELL,4:ASSET_WELL,
           5:ASSET_WELL,6:ASSET_OHT,7:ASSET_DT,8:ASSET_DT,
           9:ASSET_WELL,10:ASSET_DT,11:ASSET_DT,12:ASSET_OHT,
           13:ASSET_OHT,14:ASSET_WELL}
RMAP={1:"RCC",2:"Tiled",3:"Tin",4:"Others"}
UMAP={1:"DT",2:"OHT",3:"Well",4:"Handpump",5:"Tubewell",
      6:"Tank",7:"Solar",8:"Light",9:"Borewell",
      10:"DT",11:"DT",12:"OHT",13:"OHT",14:"Well"}

def filt(gdf, vname):
    if gdf is None or len(gdf)==0: return None
    if 'Village_Na' not in gdf.columns: return gdf
    out = gdf[gdf['Village_Na']==vname]
    return out if len(out)>0 else None

def get_util(layers, vname):
    for key in ['Utility_Poly_','Utility_Poly','Utility']:
        if key in layers:
            r = filt(layers[key], vname)
            if r is not None and len(r)>0: return r
    return None

def burn(pairs, transform, shape):
    valid=[(mapping(g),v) for g,v in pairs
           if g is not None and not g.is_empty]
    if not valid: return np.zeros(shape,dtype=np.uint8)
    return rasterize(valid,out_shape=shape,transform=transform,
                     fill=0,dtype=np.uint8,all_touched=True)

def load_all(shp_dir, names):
    out={}
    for n in names:
        p=shp_dir/f'{n}.shp'
        if p.exists():
            out[n]=gpd.read_file(str(p))
    return out

print("Loading SHP layers...")
PB=load_all(SHP_PB_DIR,['Built_Up_Area_typ','Road','Road_Centre_Line',
                         'Water_Body','Water_Body_Line','Utility_Poly_','Utility'])
CG=load_all(SHP_CG_DIR,['Built_Up_Area_type','Road','Road_Centre_Line',
                         'Water_Body','Water_Body_Line','Utility_Poly_','Utility'])
print(f"PB layers: {list(PB.keys())}")
print(f"CG layers: {list(CG.keys())}")

def process(vid):
    state,vname=VILLAGE_MAP[vid]
    layers=PB if state=='PB' else CG
    bkey='Built_Up_Area_typ' if state=='PB' else 'Built_Up_Area_type'
    vdir=TRAIN_DIR/vid
    ortho=vdir/"orthophoto.tif"
    if not ortho.exists(): print("  no orthophoto"); return
    (vdir/"masks").mkdir(exist_ok=True)
    with rasterio.open(ortho) as src:
        meta=src.meta.copy();H,W=src.height,src.width
        crs=src.crs;tr=src.transform
    print(f"  CRS={crs}  {W}x{H}")
    meta.update(count=1,dtype="uint8",compress="lzw",nodata=None)
    shape=(H,W);mask=np.zeros(shape,dtype=np.uint8)
    for lname,cls in [('Water_Body',WATER),('Road',ROAD)]:
        gdf=filt(layers.get(lname),vname)
        if gdf is not None and len(gdf)>0:
            try:
                l=burn([(g,cls) for g in gdf.to_crs(crs).geometry],tr,shape)
                mask[l>0]=cls;print(f"  {lname}: {len(gdf)}")
            except Exception as e: print(f"  {lname} err:{e}")
    for lname,buf,cls in [('Water_Body_Line',3.0,WATER),('Road_Centre_Line',None,ROAD)]:
        gdf=filt(layers.get(lname),vname)
        if gdf is not None and len(gdf)>0:
            try:
                gdf=gdf.to_crs(crs)
                if buf:
                    polys=[g.buffer(buf) for g in gdf.geometry if g and not g.is_empty]
                else:
                    polys=[]
                    for _,row in gdf.iterrows():
                        if row.geometry is None or row.geometry.is_empty: continue
                        w_=float(row.get("Width",4.0) or 4.0)
                        polys.append(row.geometry.buffer(max(2.0,min(w_,20.0))/2))
                l=burn([(g,cls) for g in polys],tr,shape)
                mask[l>0]=cls;print(f"  {lname}: {len(gdf)}")
            except Exception as e: print(f"  {lname} err:{e}")
    b=filt(layers.get(bkey),vname)
    if b is not None and len(b)>0:
        try:
            b=b.to_crs(crs)
            pairs=[]
            for _,row in b.iterrows():
                if row.geometry is None or row.geometry.is_empty: continue
                rt=int(row.get("Roof_type",4) or 4)
                pairs.append((row.geometry,ROOF_MAP.get(rt,BLDG_OTHER)))
            l=burn(pairs,tr,shape);mask[l>0]=l[l>0]
            print(f"  Buildings: {len(b)}")
            for rv,cn in [(1,"RCC"),(2,"Tiled"),(3,"Tin"),(4,"Other")]:
                cnt=(b["Roof_type"]==rv).sum()
                if cnt>0: print(f"    {cn}:{cnt}")
            with open(vdir/"roof_labels.csv","w",newline="") as f_:
                w_=csv.writer(f_)
                w_.writerow(["id","roof_type","roof_name","cx","cy","area_m2"])
                for i,row in b.iterrows():
                    c_=row.geometry.centroid;rt=int(row.get("Roof_type",4) or 4)
                    w_.writerow([i,rt,RMAP.get(rt,"Others"),
                                 round(c_.x,2),round(c_.y,2),
                                 round(float(row.get("Area_Sqm",0) or 0),2)])
        except Exception as e: print(f"  buildings err:{e}")
    u=get_util(layers,vname)
    if u is not None and len(u)>0:
        try:
            u=u.to_crs(crs);pairs=[]
            for _,row in u.iterrows():
                if row.geometry is None or row.geometry.is_empty: continue
                ut=int(row.get("Utility_Ty",1) or 1)
                g=row.geometry
                if g.geom_type=="Point": g=g.buffer(5)
                pairs.append((g,ASSET_MAP.get(ut,ASSET_DT)))
            l=burn(pairs,tr,shape);mask[l>0]=l[l>0]
            print(f"  Utilities: {len(u)}")
            with open(vdir/"asset_labels.csv","w",newline="") as f_:
                w_=csv.writer(f_)
                w_.writerow(["id","utility_ty","name","x","y"])
                for i,row in u.iterrows():
                    c_=row.geometry.centroid;ut=int(row.get("Utility_Ty",1) or 1)
                    w_.writerow([i,ut,UMAP.get(ut,"Other"),
                                 round(c_.x,2),round(c_.y,2)])
        except Exception as e: print(f"  utilities err:{e}")
    with rasterio.open(vdir/"masks"/"mask.tif","w",**meta) as dst:
        dst.write(mask,1)
    unique,counts=np.unique(mask,return_counts=True)
    total=mask.size
    cls_n=["bg","RCC","Tiled","Tin","Other","road","water","DT","OHT","Well"]
    print(f"  non-bg:{100*(1-counts[0]/total):.2f}%")
    for u_,c_ in zip(unique,counts):
        if u_>0: print(f"    {cls_n[u_]}:{c_:,}px")
    print(f"  DONE ✓")

# Check already done
done={m.split('/')[5] for m in
      glob.glob('/content/svamitva/data/train_villages/*/masks/mask.tif')}
print(f"\nAlready done: {sorted(done)}")

# Process remaining
todo=[v for v in ['village_01','village_02','village_03','village_04',
                   'village_05','village_06','village_07','village_08',
                   'village_09','village_10'] if v not in done]
print(f"Processing: {todo}\n")

for vid in todo:
    print(f"\n{'='*50}")
    print(f"Village: {vid}  ({VILLAGE_MAP[vid][1]})")
    try:
        process(vid)
    except Exception as e:
        print(f"  ERROR: {e}")
        import traceback;traceback.print_exc()

# Final summary
masks=glob.glob('/content/svamitva/data/train_villages/*/masks/mask.tif')
print(f"\n{'='*50}")
print(f"MASKS: {len(masks)}/10")
for m in sorted(masks):
    vid=m.split('/')[5]
    with rasterio.open(m) as s: data=s.read(1)
    u=np.unique(data);nb=round(100*(data>0).sum()/data.size,2)
    print(f"  {vid}: classes={u.tolist()}  non-bg={nb}%")
```
